```python
import pandas as pd

# 1) Load the files
annual_rain = pd.read_csv("viz6.csv")
monthly = pd.read_csv("chart.csv")

# 2) Rename annual rainfall columns
annual_rain.columns = ["Year", "Annual_Precip_mm"]

# 3) Rename and select relevant monthly columns
monthly = monthly.rename(columns={
    "Category": "Month",
    "Average Mean Surface Air Temperature (°C)": "Mean_Temp_C",
    "Precipitation (mm)": "Monthly_Precip_mm"
})[["Month", "Mean_Temp_C", "Monthly_Precip_mm"]]

# 4) Add numeric month order for easier merging later
month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
               "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
monthly["MonthNum"] = monthly["Month"].map({m: i+1 for i, m in enumerate(month_order)})

# 5) Ensure numeric columns are properly converted
monthly["Mean_Temp_C"] = pd.to_numeric(monthly["Mean_Temp_C"], errors="coerce")
monthly["Monthly_Precip_mm"] = pd.to_numeric(monthly["Monthly_Precip_mm"], errors="coerce")
annual_rain["Annual_Precip_mm"] = pd.to_numeric(annual_rain["Annual_Precip_mm"], errors="coerce")

# 6) Save cleaned datasets for later use
monthly.to_csv("kenya_monthly_temp_rain.csv", index=False)
annual_rain.to_csv("kenya_annual_rain.csv", index=False)

# Preview both datasets
monthly.head(), annual_rain.head()
```

```python
# Example: after uploading the WeFarm dataset that contains a 'question_date' column
# Read the dataset and parse the date column
wefarm = pd.read_csv("wefarm.csv", parse_dates=["question_date"])

# Extract year and month number from the date
wefarm["Year"] = wefarm["question_date"].dt.year
wefarm["MonthNum"] = wefarm["question_date"].dt.month

# Count number of questions per (Year, Month)
monthly_questions = (
    wefarm.groupby(["Year", "MonthNum"])
    .size()
    .reset_index(name="Questions")
)

# Merge question counts with monthly climate data
# Note: Current climate file has only typical monthly averages (no specific year)
# If yearly monthly climate data becomes available, you can replace it later.
merged = monthly_questions.merge(
    monthly[["MonthNum", "Mean_Temp_C", "Monthly_Precip_mm"]],
    on="MonthNum",
    how="left"
)

# Preview merged dataset
merged.head()
```

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Load your dataset
wefarm = pd.read_csv("wefarm.csv")

# If you’re not sure of column names:
print(wefarm.columns)
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text

wefarm["clean_text"] = wefarm["question_text"].apply(clean_text)

wefarm["clean_text"].head()
```

```python
vectorizer = CountVectorizer(stop_words='english', max_df=0.8, min_df=3)
dtm = vectorizer.fit_transform(wefarm["clean_text"])
print("✅ Vectorized shape:", dtm.shape)
```

```python
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(dtm)

# Show the top words in each topic
words = vectorizer.get_feature_names_out()
for i, topic in enumerate(lda.components_):
    top_words = [words[j] for j in topic.argsort()[-10:]]
    print(f"🧩 Topic {i+1}: {' | '.join(top_words)}")
```

```python
topic_values = lda.transform(dtm).argmax(axis=1)
wefarm["topic"] = topic_values

topic_counts = wefarm.groupby("MonthNum")["topic"].value_counts().unstack(fill_value=0)
topic_counts.head()
```

```python
# Inspect the column names first
print(wefarm.columns)
```

```python
# Replace 'question_date' with your actual date column name if different
wefarm["question_date"] = pd.to_datetime(wefarm["question_date"], errors="coerce")

# Extract month number
wefarm["MonthNum"] = wefarm["question_date"].dt.month

# Confirm
wefarm[["question_date", "MonthNum"]].head()
```

```python
topic_counts = wefarm.groupby("MonthNum")["topic"].value_counts().unstack(fill_value=0)
topic_counts.head()
```

```python
import matplotlib.pyplot as plt

# Normalize topic counts (so they’re comparable across months)
topic_trends = topic_counts.div(topic_counts.sum(axis=1), axis=0)

plt.figure(figsize=(12,6))
for col in topic_trends.columns:
    plt.plot(topic_trends.index, topic_trends[col], marker='o', label=f"Topic {col}")

plt.title("Monthly Distribution of Topics (Normalized)")
plt.xlabel("Month")
plt.ylabel("Topic Frequency (normalized)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
topic_id = 2  # choose topic index that looked like "pest" or "irrigation"

fig, ax1 = plt.subplots(figsize=(10,5))

ax1.plot(topic_counts.index, topic_counts[topic_id], color='purple', marker='o', label=f"Topic {topic_id} (Questions)")
ax1.set_xlabel("Month")
ax1.set_ylabel("Number of Questions", color='purple')

ax2 = ax1.twinx()
ax2.plot(monthly["MonthNum"], monthly["Monthly_Precip_mm"], color='blue', linestyle='--', label="Rainfall (mm)")
ax2.set_ylabel("Rainfall (mm)", color='blue')

fig.suptitle(f"Topic {topic_id} Frequency vs Rainfall", fontsize=14)
ax1.grid(True)
fig.legend(loc="upper right")
plt.show()
```
